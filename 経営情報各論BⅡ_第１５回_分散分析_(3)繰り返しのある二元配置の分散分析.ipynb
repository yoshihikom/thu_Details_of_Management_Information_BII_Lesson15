## 繰り返しのある二元配置分散分析

statsmodels で繰り返しのある二元配置の分散分析を行います。

ここでは，２つの例を示します．

１つは，歯の成長データで，もう１つは，講義で説明した例です．

```python
import pandas as pd
from scipy import stats
```

## 例１

歯の成長データを例にしています．要因はsuppとdoseとなります．
使用するデータは、<a href="https://vincentarelbundock.github.io/Rdatasets/">Rdatasets</a> から拝借しました。

```python
# 歯の成長データを取得
data = pd.read_csv('ToothGrowth.csv')
data
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
formula = 'len ~ C(supp) + C(dose) + C(supp):C(dose)'
model = ols(formula, data).fit()
print(model.summary())
```

```python
aov_table = sm.stats.anova_lm(model, typ=2)
print(aov_table)
```

$p$値を計算すると，

```python
import math
print("C(supp) pvalue", 2.311828 * pow(math.e, -4))
print("C(dose) pvalue", 4.046291 * pow(math.e, -18))
print("C(supp):C(dose) pvalue", 2.186027 * pow(math.e, -2))
```

要因suppと要因doseの$p$値は0.05（有意水準5%)以下なので棄却域に入っている．よって帰無仮説は棄却される，要因suppと要因doseの効果が確かめられた．
交互作用の$p$値は0.05以上なので棄却域に入っていない．よって帰無仮説は棄却されない．交互作用は検証できなかった．

## 例２

均等に選ばれた子豚を１か月飼育したのち体重増加を調べた資料です．因子は飼料と温度の２つです．

```python
# 子豚の体重データを取得
data = pd.read_csv('Piglets_Growth.csv')
data
```

```python
formula = 'weight ~ C(feed) + C(temp) + C(feed):C(temp)'
model = ols(formula, data).fit()
print(model.summary())
```

```python
aov_table = sm.stats.anova_lm(model, typ=2)
print(aov_table)
```

飼料と温度の効果の$p$値は0.05（有意水準5%）より小さいので，棄却域に入っているため，帰無仮説は棄却される．つまり，飼料と温度の効果が確かめられた．
交互作用の$p$値は0.05より大きいので棄却域に入っていないので，帰無仮説は棄却されません．温度と飼料が絡み合って特別な効果を生む交互作用は，この資料からは検証できなかった．
